```python
# data analysis
import pandas as pd

#data visualization
import matplotlib.pyplot as plt 
import seaborn as sns
#import plotly.express as px 
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
# import & load the data 

population= pd.read_csv('C:\\Users\\Anindita\\Pandas\\population.csv')
```

```python
# change the type of country and density value 
population['Density (P/Km²)'] = population['Density (P/Km²)'].astype(str)
population['Country'] = population['Country'].astype(str)
```

```python
# descriptive statistics

population.describe()
```

```python
# Data analysis and Visualization 

# Question 1: Top 30 countries by population in 2020

# 2020 population data 

current_population = population[population['Year'] == 2020][:30]

plt.rcParams['figure.figsize'] = (25, 7)
ax = sns.barplot(x = current_population['Country'][:30], y = current_population['Population'][:30], palette='dark')
ax.set_xlabel(xlabel = 'Countries', fontsize = 10)
ax.set_ylabel(ylabel = 'Population in Billion', fontsize = 10)
ax.set_title(label = 'Population of top 30 countries in 2020', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Question 2: Population of each country in 2020 

plt.figure(figsize=(9, 25))
ax = sns.barplot(x="Population", y="Country", data=current_population, palette="tab20c", linewidth = 1)
for i, j in enumerate(current_population["Population"]): 
    ax.text(.5, i, j, weight="bold", color='black', fontsize=10)
plt.title("Population of each country in 2020")
ax.set_xlabel(xlabel = 'Population in Billion', fontsize = 10)
ax.set_ylabel(ylabel = 'Countries', fontsize = 10)
plt.show()
```

```python
# Top 5 most populated countries in world 
population_top5_2020 = population[population['Year'] == 2020][:5]
top_5_countries = population_top5_2020['Country'].unique()

top5_population = population[population['Country'].isin(top_5_countries)][['Year','Country','Population']]
top5_population_pivot = top5_population.pivot(index='Year', columns='Country', values='Population')
top5_population_pivot.style.background_gradient(cmap='PuBu')
```

```python
# Top 5 most populated countries in the world 

fig,ax = plt.subplots(figsize=(20, 10))
sns.despine()
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth":2})

sns.barplot(x="Year", y="Population", data=top5_population, hue='Country')
ax.set_ylabel(ylabel = 'Population in Billion', fontsize = 10)
ax.set_xlabel(xlabel = 'Year', fontsize = 10)
ax.set_title('Top 5 most populated countries in the World from 1955 to 2020')
ax.legend();
```
